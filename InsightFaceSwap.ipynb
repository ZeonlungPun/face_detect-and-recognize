```python
import numpy as np
import os,cv2,glob
import matplotlib.pyplot as plt
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image as ins_get_image
```

# **detect faces**


```python
app=FaceAnalysis(name='buffalo_l')
app.prepare(ctx_id=0,det_size=(640,640))
```

```python
!ls
```

```python
img1=cv2.imread('/content/test1.jpg')
img2=cv2.imread('/content/test2.jpg')
img3=cv2.imread('/content/test3.jpg')
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
face1=app.get(img1)
face2=app.get(img2)
face3=app.get(img3)
```

```python
fig,axs=plt.subplots(1,3)
bbox1=face1[0]['bbox']
bbox1=[int(b)   for b in bbox1]
axs[0].imshow(img1[bbox1[1]:bbox1[3],bbox1[0]:bbox1[2],::-1])
axs[0].axis('off')

bbox2=face2[0]['bbox']
bbox2=[int(b)   for b in bbox2]
axs[1].imshow(img2[bbox2[1]:bbox2[3],bbox2[0]:bbox2[2],::-1])
axs[1].axis('off')

bbox3=face3[0]['bbox']
bbox3=[int(b)   for b in bbox3]
axs[2].imshow(img3[bbox3[1]:bbox3[3],bbox3[0]:bbox2[2],::-1])
axs[2].axis('off')
```

```python
swapper=insightface.model_zoo.get_model(
    '/content/drive/MyDrive/inswapper_128.onnx',
    download=False,
    download_zip=False)
```

```python
source_face=face3[0]
bbox=source_face['bbox']
bbox=[int(b) for b in bbox]
plt.imshow(img3[bbox[1]:bbox[3],bbox[0]:bbox[2],::-1])
plt.show()
```

# **swap Wong's face into Lau'face**

```python
res=img1.copy()
res=swapper.get(res,face1[0],source_face,paste_back=True)
```

# **plot the swapped face**

```python
plt.imshow(res[:,:,::-1])
plt.show()
```
